```
# Python program to scrap the website and save quetes from website
import requests # Send an HTTP request to the URL of the webpage
from bs4 import BeautifulSoup as bs # for pulling data out of HTML and XML files.
import csv
```

```
URL = "http://www.values.com/inspirational-quotes" #Specify Website which want to scrap
# Send a HTTP request to the specified URL and 
# save the response from server in a response object called r.
r = requests.get(URL)
print(r.content)
```

```
soup = bs(r.content,'html5lib')# If this line causes an error, run 'pip install html5lib' or install html5lib
# r.content = row html content ; html5lib = html parser which we want to use. 
soup.prettify()
```

```
quotes = [] # lists to store quotes
```

from soup.prettify() we notice that all the quotes are inside a div container whose id is all_quotes. So, we find that div element (termed as table in above code) using find() method

```
table = soup.find('div', attrs = {'id':'all_quotes'})
```

We create a dictionary to save all information about a quote. The nested structure can be accessed using dot notation. To access the text inside an HTML element, we use .text :

```
for row in table.findAll('div',attrs = {'class':'col-6 col-lg-3 text center margine -30px-bottom sm-margine-30x-top'}):
  quote = {}
  quote['theme']=row.h5.text
  quote['url'] = row.a['href']
  quote['img'] = row.img['src']
  quote['lines'] =row.image['alt'].split('#')[0]
  quote['author'] = row.image['alt'].split('#')[1]
  quotes.append(quote)
```

```
#we would like to save all our data in some CSV file.
filename = 'inspiration_quotes.csv'
with open(filename,'w',newline='')as f:
  w = csv.DictWriter(f,['theme','url','img','lines','author'])
  w.writeheader()
  for quote in quotes:
    w.writerrow(quote)
```
