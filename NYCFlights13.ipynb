##Q1. Write a function that given your birthday (as a date), returns how old you are in years.

```python
# Ans
from datetime import datetime

Input_Date = input("Enter your Birth Date in mm/dd/yyyy format: ")

# output of above function is string which we have to convert in datetime.
# datetime.strptime()to parse a string to datetime(convert a string to datetime)
Birth_Date = datetime.strptime(Input_Date, '%m/%d/%Y')

print("Age : %d" % ((datetime.today() - Birth_Date).days/365))

### Ans 1 Ends ###
```

##Q4. Had an delay in arrival of 3 or less hours##


```python
#Ans4

# Finding flights delayed by 3hrs or less

delayBy3hrs = flights[flights['arr_delay'] <= 3]

delayBy3hrs


#Ans4 Ends
```

```python
plt.plot(delayBy3hrs.groupby('month')['arr_delay'].mean())
plt.xlabel('Months', size = 12)
plt.ylabel('Avg Arrival delay in mins', size = 12)
plt.title('Avg Arrival Delay in mins by Month')
plt.show();
```

## Q8. Arrived more than two hours late, but didn't leave late?



```python
#Ans8
arrDelay_DepOnTime = flights.loc[(flights['arr_delay']>= 120) & (flights['dep_delay'] <= 0)]
arrDelay_DepOnTime.tail()
```

##Q9. Were delayed by at least an hour, but made up over 30 minutes in flight


```python
#Ans9

flightsMakeup30min = flights.loc[(flights['dep_delay'] >= 60) & ((flights['dep_delay'] - flights['arr_delay']) > 30)]
flightsMakeup30min.head()
```
